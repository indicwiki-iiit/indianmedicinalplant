```python
import os
import re
import selenium
import time
from selenium import webdriver
import warnings
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(executable_path="C:/Users/Suraj/Desktop/IIIT/Wiki/chromedriver_win32/chromedriver.exe")
```

```python
iucm_url = 'https://www.iucnredlist.org/'
```

```python
import json
wikiplant = {}
 
# Opening JSON file
with open('wikiplant_v1.json') as json_file:
    wikiplant = json.load(json_file)
```

```python
def get_iucn_attributes(plant_name,plant_id,wiki_dict):
    #wiki['Taxonomy'] = {}
    search_box = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div[1]/div/form/input')
    search_box.click()
    search_box.send_keys(plant_name)

    search_button = driver.find_element_by_xpath('//*[@id="nav-search"]/div/form/button')
    search_button.click()

    time.sleep(5)

    iucn_plant_id = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div/div[2]/section/div[2]/article/a')
    iucn_plant_id.click()

    time.sleep(2)

    assessment_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[2]/div[3]/div[1]')
    assessment_expand.click()

    #time.sleep(2)


    assessment = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[2]/div[3]/div[2]/div/div[2]/div/p')
    wiki_dict[plant_id]['assessment'] = assessment.text
    #print('assesment : ' , assesment.text)

    geog_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[3]/div[3]/div[1]')
    geog_expand.click()
    geog_range = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[3]/div[3]/div[2]/div/div[2]/div[1]')
    wiki_dict[plant_id]['Geography'] = geog_range.text
    #print('Geography : ' , geog_range.text)
    

    population_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[4]/div[3]/div[1]')
    population_expand.click()

    population = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[4]/div[3]/div[2]/div/div[2]/div[1]')
    population_text = population.text
    wiki_dict[plant_id]['population'] = population_text
    #print('population_text : ',population_text)


    habit_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[5]/div[3]/div[1]')
    habit_expand.click()

    habit = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[5]/div[3]/div[2]/div/div[2]/div[1]')
    habit_text = habit.text
    wiki_dict[plant_id]['iucn_habit'] = habit_text
    #print('Habit Text : ', habit_text)

    uses_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[7]/div/div[1]')
    uses_expand.click()

    uses = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[7]/div/div[2]/div/div[2]/div[1]')
    uses_text = uses.text
    wiki_dict[plant_id]['iucn_uses'] = uses_text
    #print('Uses Text : ', uses_text)


    conservation_expand = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[8]/div[3]/div[1]')
    conservation_expand.click()

    conservation = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div/div/div[5]/article[8]/div[3]/div[2]/div/div[2]/div[1]')
    conservation_text = conservation.text
    wiki_dict[plant_id]['iucn_conservation'] = conservation_text
    #print('conservation Text : ', conservation_text)
    time.sleep(2)

    return wiki_dict
```

```python
key = '229631'
warnings.filterwarnings("ignore") 
driver.get(iucm_url)
time.sleep(2)
wiki_ind = wikiplant[key]
plant_name = wiki_ind['Name']
#wikiplant[key]['assessment'] = 'test'
wikiplant = get_iucn_attributes(plant_name,key,wikiplant)
```

```python
count = 0
warnings.filterwarnings("ignore")
for key in wikiplant.keys():
    driver.get(iucm_url)
    time.sleep(2)
    count = count+1
    if count%50==0:
        print('Processed ', count)
    wiki_ind = wikiplant[key]
    plant_name = wiki_ind['Name']
    try:
        wikiplant = get_iucn_attributes(plant_name,key,wikiplant)
    except:
        continue
    
    
```
