```python
import os
import re
import selenium
import time
from selenium import webdriver
import warnings
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(executable_path="C:/Users/Suraj/Desktop/IIIT/Wiki/chromedriver_win32/chromedriver.exe")
```

```python
wikiplant = {}
warnings.filterwarnings("ignore") 
count = 0
os.chdir('C:\\Users\\Suraj\\Desktop\\IIIT\\Wiki')
bio = open('Webpage_element_text.txt','r')
bio_text= bio.read()
#print(bio_text)

regex = 'href="/group/medicinal_plants/species/show/(.*?)">'
plant_number_list = re.findall(regex,bio_text)
print(len(plant_number_list))

for p in plant_number_list:
    count = count + 1
    url = 'https://indiabiodiversity.org/group/medicinal_plants/species/show/' + p
    if count%50 == 0:
        print('processed ',count)
    print(url)
    wiki = get_scrapped_dict(url)
    wikiplant[p] = wiki
```

```python
def get_scrapped_dict(URL):
    driver.get(URL)
    time.sleep(2)
    # Get Plant Name
    wiki = {}
    plant_name = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[1]/div[1]/div[2]/div[1]/h1/span/i')
    wiki['Name'] = plant_name.text

    ##Taxonomy Info
    wiki['Taxonomy'] = {}
    #taxonomy_table = driver.find_element_by_xpath('//table[@class = "chakra-table css-5605sr"]')
    taxonomy_table = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[2]/div[1]/div[2]/table')
    for row in taxonomy_table.find_elements_by_xpath('.//tr'):
        row_i = row.find_elements_by_tag_name('td')
        wiki['Taxonomy'][row_i[0].text] = row_i[1].text

    #Synonyms
    synonyms = ''
    try:
        synonym_table = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[2]/div[2]/div/div/table')
        for row in synonym_table.find_elements_by_xpath('.//tr'):
            row_i = row.find_elements_by_tag_name('td')
            synonyms = synonyms + ' , ' + row_i[1].text
        wiki['Synonym'] = synonyms[3:]
    except:
        wiki['Synonym'] = ''

    ##Common Names
    wiki['Common Name'] = {}
    try:
        synonym_table = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[3]/div[2]/div/div/table')
        for row in synonym_table.find_elements_by_xpath('.//tr'):
            row_i = row.find_elements_by_tag_name('td')
            wiki['Common Name'][row_i[0].text] = row_i[1].text
    except:
        wiki['Common Name'] = {}
    
    try:
        brief = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[1]/div[2]/div/div/div[3]/div/div/div/div[1]')
        wiki['Brief'] = brief.text
    except:
        wiki['Brief'] = ''
    #time.sleep(1)

    try:
        summary = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[1]/div[2]/div/div/div[2]/div/div/div/div[1]')
        wiki['Summary'] = summary.text
    except:
        wiki['Summary'] = ''
    #time.sleep(1)

    try:
        diag_descr = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[1]/div[2]/div/div/div[4]/div[3]/div/div/div[1]/div[1]')
        wiki['Diagnostic Description1'] = diag_descr.text
    except:
        wiki['Diagnostic Description1'] = ''

    try:
        Plant_descr = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[1]/div[2]/div/div/div[4]/div[3]/div/div/div[2]/div[1]')
        wiki['Plant Descr'] = Plant_descr.text
    except:
        wiki['Plant Descr'] = ''

    wiki['Diagnostic Description2'] = {}
    try:
        d1 = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[1]/div[2]/div/div/div[4]/div[3]/div/div/div[2]/div[2]')
        h6_tags = d1.find_elements_by_tag_name('h6')
        p_tags = d1.find_elements_by_tag_name('p')

        for i in range(len(h6_tags)):
            wiki['Diagnostic Description2'][h6_tags[i].text] = p_tags[i].text
            #print(h6_tags[i].text, ':', p_tags[i].text)
    except:
        wiki['Diagnostic Description2'] = {}

    ## Morphology Cyclicity - Season/Month when it flowers/fruits ##
    try:
        cyclicity_1 = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[3]/div/div/div[1]/div[1]')
        cyclicity_2 = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[3]/div/div/div[2]/div[1]')

        wiki['cyclicity_1'] = cyclicity_1.text
        wiki['cyclicity_2'] = cyclicity_2.text
    
    except:
        wiki['cyclicity_1'] = ''
        wiki['cyclicity_2'] = ''

    
    ## Morphology Tags ##

    wiki['Morphology'] = {}
    try:
        growth_form_value = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[10]/div/div/div[1]/div/div/div')
        wiki['Morphology']['Growth Form'] = growth_form_value.text
    except:
        wiki['Morphology'] = {}

    try:
        morph_fields = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[10]/div/div/div[2]/div[2]')
        h6_tags = morph_fields.find_elements_by_tag_name('h6')
        p_tags = morph_fields.find_elements_by_tag_name('p')

        for i in range(len(h6_tags)):
            wiki['Morphology'][h6_tags[i].text] = p_tags[i].text
            #print(h6_tags[i].text, ':', p_tags[i].text)
    except:
        wiki['Morphology'] = {}

    ## Morphology General Description ##
    try:
        morphology_text = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[10]/div/div/div[3]/div[1]')
        wiki['Morphology']['Descr'] = morphology_text.text
    except:
        wiki['Morphology']['Descr'] = ''

    ## Ecology ##
    try:
        ecology = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[3]/div[2]/div/div/div[13]/div/div/div/div[1]')
        wiki['Ecology'] = ecology.text
    except:
        wiki['Ecology'] = ''

    try:
        habitat = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[4]/div[2]/div/div/div[2]/div/div/div/div[1]')
        wiki['Habitat'] = habitat.text
    except:
        wiki['Habitat'] = ''
    
    #Distribution Details
    try:
        distribution_general = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[4]/div[2]/div/div/div[3]/div[2]/div/div/div[1]/div[1]')
        wiki['General Distribution'] = distribution_general.text
    except:
        wiki['General Distribution'] = ''

    # Get Global and Indian Distribution
    try:
        distribution_element = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[4]/div[2]/div/div/div[3]/div[2]/div/div/div[2]/div[1]')
        h6_tags = distribution_element.find_elements_by_tag_name('h6')
        p_tags = distribution_element.find_elements_by_tag_name('p')

        for i in range(len(h6_tags)):
            wiki[h6_tags[i].text] = p_tags[i].text
    except:
        wiki['Global Distribution'] = ''
        wiki['Indian Distribution'] = ''
    

    try:
        Uses = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[7]/div[2]/div/div/div[2]/div/div/div[3]/div[1]/p')
        wiki['Uses'] = Uses.text
    except:
        wiki['Uses'] = ''
    
    try:
        gereral_uses = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[7]/div[2]/div/div/div[2]/div/div/div[4]/div[1]')
        wiki['Uses General'] = gereral_uses.text
    except:
        wiki['Uses General'] = ''

    try:
        folklore = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div[4]/div[7]/div[2]/div/div/div[3]/div/div/div/div[1]')
        wiki['Folklore'] = folklore.text
    except:
        wiki['Folklore'] = ''

    try:
        try:
            references = driver.find_element_by_xpath('//*[@id="information-listing"]/div[2]/div/div/div[3]/div[3]/ol')
        except:
            references = driver.find_element_by_xpath('//*[@id="information-listing"]/div[2]/div/div/div[3]/div[2]/ol')
        ref_list = references.find_elements_by_tag_name('li')
        refer_list = []
        for r in ref_list:
            refer_list.append(r.text)
        wiki['References'] = refer_list
    except:
        wiki['References'] = []
    
    wiki['image_files'] = {}
    all_media_button = driver.find_element_by_xpath('//*[@id="main"]/div/div[1]/div[2]/div/div[1]/button')
    all_media_button.click()
    time.sleep(2)
    try:
        img_boxes = driver.find_elements_by_xpath("//img[contains(@class,'chakra-image css-1tod0jh')]")
        for i in range(len(img_boxes)):
            wiki['image_files'][i] = img_boxes[i].get_attribute('src')
    except:
        wiki['image_files'] = {}
        
        
    return wiki
```

```python
import json
wikiplant = {}
# Opening JSON file
with open('wikiplant_v2.json') as json_file:
    wikiplant = json.load(json_file)

wikicommons_url = 'https://commons.wikimedia.org/'
warnings.filterwarnings("ignore") 

def get_image_file_names(plant_name):
    driver.get(wikicommons_url)
    time.sleep(2)
    search_box = driver.find_element_by_xpath('/html/body/div[4]/div[1]/div[2]/div/div/form/div/input[1]')
    search_box.send_keys(plant_name)

    search_button = driver.find_element_by_xpath('//*[@id="searchButton"]')
    search_button.click()
    time.sleep(2)

    image_list = driver.find_elements_by_class_name("sdms-image-result")
    count = 0
    img_list = []
    for img in image_list:
        img_list.append(img.get_attribute('title').split('File:')[1])
        count = count + 1
        
        if count == 10:
            break
    
    return img_list
```
